```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper = Nitter()
```

```python
tweets = scraper.get_tweets('japanearthquake', mode = 'hashtag', number = 20)
```

```python
tweets
```

```python
final_tweets = []
for x in tweets['tweets']:
    data = [x['link'], x['text'],x['date'],x['stats']['likes'],x['stats']['comments']]
    final_tweets.append(data)
```

```python
df = pd.DataFrame(final_tweets, columns =['twitter_link','tweet_text','timestamp','likes','comments'])
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# data = {
#     'timestamp': ["Feb 8, 2024 · 12:20 PM UTC", "Feb 8, 2024 · 1:45 PM UTC", "Feb 8, 2024 · 2:10 PM UTC", "Feb 8, 2024 · 3:30 PM UTC"],
#     'tweet_text': ["Tweet 1", "Tweet 2", "Tweet 3", "Tweet 4"]
# }

# # Create DataFrame
# df = pd.DataFrame(data)

df['timestamp'] = pd.to_datetime(df['timestamp'], format='%b %d, %Y · %I:%M %p UTC')

df['hour'] = df['timestamp'].dt.hour
df['date'] = df['timestamp'].dt.date

hourly_tweets = df.groupby('hour').size()

daily_tweets = df.groupby('date').size()

plt.figure(figsize=(10, 6))
hourly_tweets.plot(kind='bar')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Tweets')
plt.title('Number of Tweets vs. Hour of the Day')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(10, 6))
daily_tweets.plot(kind='line', marker='o')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.title('Number of Tweets vs. Date')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_tweets_frequency(df, date_column):
    df[date_column] = pd.to_datetime(df[date_column])
    
    df.set_index(date_column, inplace=True)
    
    tweets_per_minute = df.resample('T').size()
    tweets_per_hour = df.resample('H').size()
    tweets_per_day = df.resample('D').size()
    
    fig, axs = plt.subplots(3, 1, figsize=(20, 10))

    axs[0].plot(tweets_per_minute)
    axs[0].set_title('Number of Tweets per Minute')
    axs[0].set_xlabel('Time')
    axs[0].set_ylabel('Number of Tweets')
    axs[0].grid(True)

    axs[1].plot(tweets_per_hour)
    axs[1].set_title('Number of Tweets per Hour')
    axs[1].set_xlabel('Time')
    axs[1].set_ylabel('Number of Tweets')
    axs[1].grid(True)

    axs[2].plot(tweets_per_day)
    axs[2].set_title('Number of Tweets per Day')
    axs[2].set_xlabel('Time')
    axs[2].set_ylabel('Number of Tweets')
    axs[2].grid(True)

    plt.tight_layout()
    plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

def show_wordcloud(data, mask=None, title=""):
    text = " ".join(t for t in data.dropna())
    stopwords = set(STOPWORDS)
    stopwords.update(["t", "co", "https", "amp", "U", "Comment", "text", "attr", "object"])
    wordcloud = WordCloud(stopwords=stopwords, scale=4, max_font_size=50, max_words=500,mask=mask, background_color="white").generate(text)
    fig = plt.figure(1, figsize=(16,16))
    plt.axis('off')
    fig.suptitle(title, fontsize=20)
    fig.subplots_adjust(top=2.3)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.show() 
```

```python
turkey = pd.read_csv('data/turkey.csv')
```

```python
turkey
```

```python
plot_tweets_frequency(turkey, 'Tweet Posted Time')
```

```python
show_wordcloud(turkey['Tweet Content'], title = 'Prevalent words in tweets')
```
