```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('0f_turkey.csv')

# Convert the 'Tweet Posted Time' to datetime
df['Tweet Posted Time'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S%z')

# Set 'Tweet Posted Time' as the index
df.set_index('Tweet Posted Time', inplace=True)

# Resample the data to get tweet counts
tweets_per_minute = df.resample('T').size()
tweets_per_hour = df.resample('H').size()
tweets_per_day = df.resample('D').size()

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(12, 10))

# Customize color palette and style
colors = ['#FF5733', '#33FF57', '#3357FF']  # Custom colors for each plot
plt.style.use('seaborn-v0_8-ticks')  # Using a dark theme for better aesthetics

# Plot number of tweets per minute
axs[0].plot(tweets_per_minute, color=colors[0], lw=1)
axs[0].set_title('Number of Tweets per Minute', fontsize=14, weight='bold')
axs[0].set_xlabel('Time', fontsize=12, weight='bold')
axs[0].set_ylabel('Number of Tweets', fontsize=12, weight='bold')
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)
axs[0].spines['left'].set_linewidth(1.5)
axs[0].spines['bottom'].set_linewidth(1.5)
axs[0].grid(False)  # Remove grid lines

# Plot number of tweets per hour
axs[1].plot(tweets_per_hour, color=colors[1], lw=2)
axs[1].set_title('Number of Tweets per Hour', fontsize=14, weight='bold')
axs[1].set_xlabel('Time', fontsize=12, weight='bold')
axs[1].set_ylabel('Number of Tweets', fontsize=12, weight='bold')
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
axs[1].spines['left'].set_linewidth(1.5)
axs[1].spines['bottom'].set_linewidth(1.5)
axs[1].grid(False)  # Remove grid lines

# Plot number of tweets per day
axs[2].plot(tweets_per_day, color=colors[2], lw=2)
axs[2].set_title('Number of Tweets per Day', fontsize=14, weight='bold')
axs[2].set_xlabel('Time', fontsize=12, weight='bold')
axs[2].set_ylabel('Number of Tweets', fontsize=12, weight='bold')
axs[2].spines['top'].set_visible(False)
axs[2].spines['right'].set_visible(False)
axs[2].spines['left'].set_linewidth(1.5)
axs[2].spines['bottom'].set_linewidth(1.5)
axs[2].grid(False)  # Remove grid lines

# Adjust layout and spacing
plt.tight_layout(pad=2)
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Assuming you have all your data to plot
data_to_plot = [tweets_per_day3, tweets_per_day2, tweets_per_day1, tweets_per_day, tweets_per_minute]
labels = ['Mexico', 'Iraq Iran', 'Syria', 'Turkey', 'Tweets per Minute']
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Create a figure object
p = figure(title='Number of Tweets per Day', x_axis_label='Time', y_axis_label='Number of Tweets')

# Iterate over each dataset and plot it
for data, label, color in zip(data_to_plot, labels, colors):
    source = ColumnDataSource(data=dict(x=data.index, y=data.values))
    p.line(x='x', y='y', source=source, legend_label=label, line_color=color)

# Add legend
p.legend.location = 'top_left'

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Category10

# Assuming you have all your data to plot
data_to_plot = [tweets_per_day3, tweets_per_day2, tweets_per_day1, tweets_per_day, tweets_per_minute]
labels = ['Mexico', 'Iraq Iran', 'Syria', 'Turkey', 'Tweets per Minute']

# Create a figure object
p = figure(title='Number of Tweets per Day', x_axis_label='Time', y_axis_label='Number of Tweets')

# Define a palette for coloring lines
palette = Category10[10]

# Iterate over each dataset and plot it
for i, (data, label) in enumerate(zip(data_to_plot, labels)):
    source = ColumnDataSource(data=dict(x=data.index, y=data.values))
    line = p.line(x='x', y='y', source=source, line_width=2, color=palette[i % 10], legend_label=label)
    # Add hover tooltips
    hover_tool = HoverTool(renderers=[line], tooltips=[('Time', '@x{%F}'), ('Tweets', '@y')], formatters={'@x': 'datetime'})
    p.add_tools(hover_tool)

# Add legend
p.legend.location = 'top_left'
p.legend.click_policy = 'hide'

# Customize plot styling
p.title.text_font_size = '16pt'
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_font_size = '14pt'
p.legend.label_text_font_size = '12pt'

# Show the plot
show(p)
```
